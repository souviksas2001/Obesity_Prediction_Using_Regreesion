```python
def NObeyesdad_Labels (n):
  if n == 'Normal_Weight':
    return 0
  else:
    return 1
```

```python
df['NObeyesdad'] = df['NObeyesdad'].apply(NObeyesdad_Labels).astype(int)
```

```python
df['family_history_with_overweight'] = df['family_history_with_overweight'].apply(lambda x : 1 if x == 'yes' else 0)
```

```python
df['FAVC'] = df['FAVC'].apply(lambda x : 1 if x == 'yes' else 0)
```

```python
df['CAEC'].value_counts()
```

```python
df['SMOKE'] = df['SMOKE'].apply(lambda x : 1 if x == 'yes' else 0)
```

```python
def CAEC_labels (n):
  if n == "Sometimes":
    return 1
  elif n == "Frequently":
    return 0
  elif n == "Always":
    return 2
  else:
    return 3
```

```python
df['CAEC'] = df['CAEC'].apply(CAEC_labels)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression

x = df['Weight']
y = df['NObeyesdad']
x = x.values.reshape(-1, 1)
model = LogisticRegression()
model.fit(x, y)

x_values = np.linspace(min(x), max(x), 1000).reshape(-1, 1)
y_probabilities = model.predict_proba(x_values)[:, 1]

sns.scatterplot(x='Weight', y='NObeyesdad', data=df, label='Data')
plt.plot(x_values, y_probabilities, color='red', label='Logistic Regression Curve')


plt.xlabel('Weight')
plt.ylabel('NObeyesdad')
plt.title('Logistic Regression Curve')


plt.legend()
plt.show()
```
